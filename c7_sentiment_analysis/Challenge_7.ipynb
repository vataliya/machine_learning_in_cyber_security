```python
import re
import string
import time

def add_to_dict(string):
  if string in word_dict.values():
    pass
  else:
    x=len(word_dict) 
    word_dict[string] = x
  return word_dict
```

```python
url_list=[]
def get_urls(string):
  urls = re.findall('(?:(?:https?|ftp):\/\/)?[\w/\-?=%.]+\.[\w/\-?=%.]+', string)
  for url in urls:
    if url.find('http:') != -1:
      url_list.append(url)
      add_to_dict(url)
    elif url.find('https:') != -1:
      url_list.append(url)
      add_to_dict(url)
    else:
      pass
  return url_list
```

```python
results = []
for line in text:
  
    clean_string = line.replace('\n','')
    clean_string = clean_string.rstrip(",.")
    ListOfWords = re.split(" ", clean_string)

    results.append(" ".join(word.lower() for word in ListOfWords))
```

```python
punct = '!"$\'()*+,-;<>[\]^_`{|}~'
clean_results=[]

for line in results:
  # print(line)
  line = line.encode('ascii',errors='ignore')
  line = line.decode()
  line = "".join([word for word in line if word not in punct])
  ListOfWords = re.split(" ", line)
  for words in ListOfWords:
    if words not in word_dict.keys():
      add_to_dict(words)
      print(len(word_dict))
  clean_results.append(line)
```

```python
input_list = []
for i in range(0,len(clean_results)):
    line_list=[]
    for word in clean_results[i].split(" "):
        line_list.append(word_dict[word])
        line_list = list(set(line_list))
        line_list.sort()
    input_list.append(' '.join([str(elem) for elem in line_list]))
```

```python
import csv
with open('word_dict_4.csv', 'w') as f:
    for key in word_dict.keys():
        f.write("%s,%s\n"%(key,word_dict[key]))
```

```python
reader = csv.reader(open('word_dict_4.csv', 'r'))
load_dict = {}
for row in reader:
   k, v = row
   load_dict[k] = v
```

```python
input_list = []
for i in range(0,len(clean_results)):
    line_list=[]
    for word in clean_results[i].split(" "):
        line_list.append(load_dict[word])
        line_list = list(set(line_list))
        line_list.sort()
    input_list.append(' '.join([str(elem) for elem in line_list]))
```

```python
with open('input_list.txt', 'w') as f:
    for item in input_list:
        f.write("%s\n" % item)
```
