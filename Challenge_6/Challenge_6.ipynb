## Challenge 6

Import required libraries and load the dataset from ARFF filetype using Scipy

```python
from scipy.io import arff
import pandas as pd
import numpy as np

data = arff.loadarff('dataset_6.arff')
df = pd.DataFrame(data[0])

df.head()
```

```python
df.info()
```

Initiate Isolation Forest to identify outliers:

```python
from sklearn.ensemble import IsolationForest
isoforest = IsolationForest(random_state=0) # Random_state used to regenerate same results for every runtime.
isoforest.fit(df)
```

```python
df[isoforest.predict(df) == -1]
```

This models identifies 12313 outliers, which is about 12.7% of the dataset. <br>
Now, identify the outliers with decision score less than -0.2:

```python
df[isoforest.decision_function(df) <= -0.2]
```

Implement Local Outlier Factor to identify outliers in the dataset with number of neighbors,k, set to 3:

```python
from sklearn.neighbors import LocalOutlierFactor
lof_model = LocalOutlierFactor(n_neighbors=3)
lof_model.fit(df)
```

```python
df[lof_model.fit_predict(df) == -1]
```

This model identifies 10564 outliers from the dataset, which is about 10.8% of the dataset.<br>
Now, identify the records with Negative Outlier Factor <= -40 :

```python
df[lof_model.negative_outlier_factor_ <=-40]
```
